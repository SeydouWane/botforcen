```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse

BASE_URL = "https://preprod2.force-n.sn"
```

```python


def get_all_links(base_url):
    visited = set()
    to_visit = [base_url]
    all_links = set()

    while to_visit:
        current_url = to_visit.pop(0)
        if current_url in visited:
            continue

        print(f"{current_url}")
        visited.add(current_url)

        try:
            res = requests.get(current_url, timeout=10)
            if res.status_code != 200:
                continue
            soup = BeautifulSoup(res.text, 'html.parser')
            for a in soup.find_all('a', href=True):
                href = a['href']
                href = urljoin(base_url, href)  

                
                if href.startswith(base_url) and not any(x in href for x in ['#', '.pdf', '.jpg', '.png', '.jpeg', '/user', '/admin']):
                    if href not in all_links:
                        all_links.add(href)
                        to_visit.append(href)
        except Exception as e:
            print(f"[Erreur] {e}")
            continue

    return sorted(all_links)


```

```python
if __name__ == "__main__":
    links = get_all_links(BASE_URL)

    print("\n Liste des URLs extraites :\n")
    for url in links:
        print(f'"{url}",')
```
